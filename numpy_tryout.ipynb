```python
import numpy as np
from NeuralNetwork import ReLU
from NeuralNetwork import StdLogisticFunction
from NeuralNetwork import RandomInitializer
from NeuralNetwork import FixedInitializer
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6]).reshape(-1, 1)
print(a.shape)
print(b.shape)
c = a + b
print(c)
print(c.shape)

print(type(a))
print(type(b))
print(type(c))
```

```python
a = np.random.uniform(low=-0.5, high=0.5, size=(2, 5))
b = np.zeros(6).reshape(-1, 1) # Reshape to make it a column vector. I think this makes it more robust to matrix operations.

print(a.shape)
print(b.shape)
```

```python
activation = StdLogisticFunction()
a = activation.apply(2)
b = activation.apply([1, 2])
c = activation.apply([[-1, 0], [1, 2]])
d = activation.apply([[1], [2], [3]])

print(a)
print(b)
print(c)
print(d)

print(a.shape)
print(b.shape)

print(type(a))
print(type(b))
print(type(c))
```

```python
a = np.array([1, 2, 3]).reshape(-1, 1)
b = a.transpose()
c = np.array(1)
d = c.reshape(-1, 1)

print(a)
print(a.shape)
print(b)
print(b.shape)
print(c)
print(c.shape)
print(type(c))
print(d)
print(d.shape)
print(d.ravel())
```

```python
initializer = RandomInitializer()

sizes = [(1, 1), (2, 1), (2, 2)]

for s in sizes:
    input_size, output_size = s
    weights, offsets = initializer.init(input_size=input_size, output_size=output_size)
    print("size:", s)
    print(weights, "\n\n", offsets, "\n$")
    print(weights.shape, "\n", offsets.shape, "\n$")
    print("", type(weights), "\n", type(offsets))
    print("\n***\n")
```

```python
initializer = FixedInitializer(weights=([[0.2, 0.5],
                                         [0.3, 0.6],
                                         [0.4, 0.7]],
                                        [[0.6,
                                          0.7]]),
                               offsets=([0.1, 0.2, 0.3], [0.4]))

weights, offsets = initializer.init(input_size=2, output_size=3)
print(weights, "\n\n", offsets, "\n$")
print(weights.shape, "\n", offsets.shape, "\n$")
print("", type(weights), "\n", type(offsets))
print("\n***\n")
```

```python
initializer = FixedInitializer(weights=2,
                               offsets=3)

weights, offsets = initializer.init(input_size=1, output_size=1)
print(weights, "\n\n", offsets, "\n$")
print(weights.shape, "\n", offsets.shape, "\n$")
print("", type(weights), "\n", type(offsets))
print("\n***\n")
```
