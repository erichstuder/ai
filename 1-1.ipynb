```python
import numpy as np
import matplotlib.pyplot as plt

from NeuralNetwork import NeuralNetwork
from NeuralNetwork import RandomInitializer
from NeuralNetwork import StdLogisticFunction
```

```python
net = NeuralNetwork(
    neurons_per_layer=[1],
    weight_initializer=RandomInitializer(),
    activation_function=StdLogisticFunction())
```

```python
net.query([0.8])
```

```python
x = np.arange(-20, 20, 0.01)
y = [net.query([val])[0] for val in x]

plt.plot(x, y)
plt.xlabel('Input (x)')
plt.ylabel('Output (y)')
plt.title('Neural Network Output vs Input')
plt.yticks(np.arange(0, 1.1, 0.1))
plt.grid(True)
plt.show()
```
