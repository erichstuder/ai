```python
import sys
sys.path.append("..")

import numpy as np
import matplotlib.pyplot as plt

from NeuralNetwork import NeuralNetwork
from NeuralNetwork import FixedInitializer
from NeuralNetwork import StdLogisticFunction
from NeuralNetwork import ReLU
```

```python
net1 = NeuralNetwork(
    neurons_per_layer=[1, 3, 1],
    weight_initializer=FixedInitializer(([[1],[1],[0]],[[3,3,1]]), [[-1,-2,1],[0]]),
    activation_function=ReLU())

net2 = NeuralNetwork(
    neurons_per_layer=[1, 3, 1],
    weight_initializer=FixedInitializer(([[1],[-0.83],[-0.8]],[[4.09, -2.37, -1.44]]), [[-1.97, 2.53, 0.71],[-1.38]]),
    activation_function=StdLogisticFunction())

net1.pretty_print_weights()
```

```python
print(net1.query(1))
print(net1.query(2))
print(net1.query(4))

print(net2.query(1)*20)
print(net2.query(2)*20)
print(net2.query(4)*20)
```

```python
d = 0.1
x = np.arange(0, 5+d, d)
y = x * x

y_net1 = []
y_net2 = []
for i in x:
    y_net1.append(net1.query(i))
    y_net2.append(20 * net2.query(i))


plt.plot(x, y)
plt.plot(x, y_net1)
plt.plot(x, y_net2)
plt.legend(["y = x^2", "net1 with ReLU", "net2 with StdLogisticFunction"])

plt.xlabel("x")
plt.ylabel("y")
plt.title("y = x^2")
plt.grid(True)
plt.show()
```
